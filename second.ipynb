```python
from sklearn.cluster import KMeans
from skimage import segmentation, color
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
import numpy as np

# Load the image
img = imread('first.jpg')

imshow(img)
plt.show()
# Convert the image to LAB color space
img_lab = color.rgb2lab(img)
# print(img)

# Reshape the image to a 2D array of pixels
rows, cols, ch = img_lab.shape
img_2d = img_lab.reshape(rows*cols, ch)

# Perform k-means clustering on the image pixels
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(img_2d)
# Reshape the k-means labels to the original image shape
superpixels = kmeans.labels_.reshape(rows, cols)
print(superpixels)
imshow(color.label2rgb(superpixels,img))
plt.show()
```
